```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline

model_name = "distilbert-base-uncased-distilled-squad"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForQuestionAnswering.from_pretrained(model_name)

qa_pipeline = pipeline("question-answering", model=model, tokenizer=tokenizer)
```

```python
def compute_em_f1(pred_answer, true_answer):
    em = int(pred_answer.strip().lower() == true_answer.strip().lower())

    pred_tokens = pred_answer.strip().lower().split()
    true_tokens = true_answer.strip().lower().split()
    common = set(pred_tokens) & set(true_tokens)
    
    if len(common) == 0:
        f1 = 0
    else:
        precision = len(common) / len(pred_tokens)
        recall = len(common) / len(true_tokens)
        f1 = 2 * (precision * recall) / (precision + recall)
    
    return em, f1
```

```python
em_total, f1_total = 0, 0

num_samples = len(df)  # <- here your dataset is used

for i in range(num_samples):
    row = df.iloc[i] 
    context = row['context']   
    question = row['question']
    true_answer = row['answer']

    result = qa_pipeline(question=question, context=context)
    pred_answer = result['answer']

    em, f1 = compute_em_f1(pred_answer, true_answer)
    em_total += em
    f1_total += f1

avg_em = em_total / num_samples
avg_f1 = f1_total / num_samples

print(f"Exact Match (EM): {avg_em:.2f}")
print(f"F1 Score: {avg_f1:.2f}")
```
